```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path

root_folder = exp_path = Path().absolute() / "experiments" 

subfolder_list = [f for f in root_folder.iterdir()]
subfolder_list_creation_time = [f.stat().st_ctime for f in subfolder_list]
latest_subfolder_idx =  subfolder_list_creation_time.index(max(subfolder_list_creation_time))

experiment_path = subfolder_list[latest_subfolder_idx]
df = pd.read_csv(experiment_path / "results.csv")
```

```python
print('Random state', df.random_state.unique())
print('Scenario id', df.scenario_id.unique())

df.info()
df.head(10)
```

```python
df_synthesis = df.groupby(["scenario_id", "random_state", "contributivity_method", "minibatch_count"], as_index=False).first()
df_synthesis = df_synthesis.drop(["amount_per_node", "contributivity_score", "contributivity_std", "node_id"], axis='columns')
df_synthesis.head()
```

```python
df_synthesis.short_scenario_name = df_synthesis.samples_split_option + df_synthesis.amounts_per_node + ' mb ' + df_synthesis.minibatch_count.astype(str)
df_synthesis.short_scenario_name.unique()
```

```python
for y_data in ["federated_test_score", "federated_computation_time"]:

    
    for samples_split_option in df_synthesis.samples_split_option.unique():

        df_plot = df_synthesis[df_synthesis.samples_split_option == samples_split_option]
        plt.figure(figsize=(10, 6), dpi=120)
        plt.title(samples_split_option + ' - ' + y_data)
        ax = sns.boxplot(x="minibatch_count", y=y_data,
                         data=df_plot, linewidth=2.5)
```

```python
def get_x_tick_amount(df, scenario_id, node_ids):
    
    x_ticks = []
    for node_id in node_ids:
        current_df = df[(df.scenario_id == scenario_id) & (df.node_id == node_id)]
        amount = current_df.amount_per_node.unique()

        assert(len(amount) == 1)
        amount = amount[0]
    
        x_ticks.append('node_' + str(node_id) + ' ' + str(amount))
    return x_ticks
```
